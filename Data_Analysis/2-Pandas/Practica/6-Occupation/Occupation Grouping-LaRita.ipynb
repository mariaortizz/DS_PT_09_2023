# Occupation

### Paso 1. Importar librerias

```python
import pandas as pd 
```

### Paso 2. Importar el conjunto de datos de esta [dirección](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Paso 3. Asígnalo a una variable llamada usuarios.

```python
usuarios = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep = '|')
usuarios
```

### Paso 4. Descubre cuál es la edad media por ocupación

```python
usuarios.groupby('occupation').mean(numeric_only = True).drop(['user_id'], axis=1)
```

### Paso 5. Descubra la proporción de Hombres por ocupación y ordénela de mayor a menor

#### Forma 1

```python
df_M_ = (usuarios[['occupation','gender']][usuarios['gender'] == 'M'].groupby('occupation').count()/ usuarios.groupby('occupation').count()[['gender']]).sort_values(by='gender', ascending=False).rename(columns={'gender':'proporcion_H'})
df_M_
```

#### Forma 2

```python
conteo_genero = usuarios.groupby('occupation')['gender'].value_counts(normalize=True).unstack()
conteo_genero[['M']].sort_values(by='M',ascending=False)
```

#### Forma 3

```python
tabla_p = usuarios.pivot_table(index = 'occupation', columns = 'gender', aggfunc = 'size')
pd.DataFrame((tabla_p['M'] / tabla_p.sum(axis=1)).sort_values(ascending=False)).rename(columns={0:'M'})
```

### Paso 6. Para cada ocupación, calcule las edades mínima y máxima

```python
usuarios.groupby('occupation')[['age']].aggregate(['min','mean','max'])
```

### Paso 7. Para cada combinación de ocupación y sexo, calcule la edad media

```python
usuarios.groupby(['occupation', 'gender']).mean(numeric_only=True).drop(['user_id'],axis = 1)
```

### Paso 8.  Para cada ocupación, presente el porcentaje de mujeres y hombres

```python
df_F = (usuarios[['occupation','gender']][usuarios['gender'] == 'F'].groupby('occupation').count() / usuarios.groupby('occupation').count()[['gender']]).sort_values(by='occupation').reset_index().fillna(0.00)
df_M = (usuarios[['occupation','gender']][usuarios['gender'] == 'M'].groupby('occupation').count() / usuarios.groupby('occupation').count()[['gender']]).sort_values(by='occupation').reset_index()
df_final = pd.merge(df_F,df_M, on='occupation').rename(columns={'gender_x':'gender_F', 'gender_y':'gender_M'})
df_final['total'] = (df_final['gender_F']) + df_final['gender_M']

df_final
```
