## **Librerias**

```python
import requests
import datetime
import hashlib
import pandas as pd
import time
```

## **Crear DataFrame** atacando la `API`

```python
def has_params(timestamp,priv_key,pub_key):
    hash_md5 = hashlib.md5()
    hash_md5.update(f'{timestamp}{priv_key}{pub_key}'.encode('utf-8'))
    hashed_params = hash_md5.hexdigest()
    return hashed_params
```

```python
# timestamp = datetime.datetime.now().strftime('%Y-%m-%d%H:%M:%S')
timestamp = str(int(time.time()))
pub_key = 'a5ed705716f74b791f28f8fc4f5239f2'
priv_key = 'a161c69e571438d578823e3ebc065a65cc61b16e'
```

```python
params = {'limit' : 100,
          'ts' : timestamp,
          'nameStartsWith' :'M',
          'apikey' : pub_key,
          'hash' : has_params(timestamp,priv_key,pub_key)}
```

```python
print(params)
```

```python
url = 'http://gateway.marvel.com/v1/public/characters'
respuesta = requests.get(url,params)

print(params)
```
